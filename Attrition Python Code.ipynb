```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data
df = pd.read_csv('/content/greendestination (1).csv')
```

```python
df.shape    # 1470 rows, 35 cols
df.duplicated().sum()   # 0 duplicates
df.isna().sum()   # 0 null values
```

```python
import sys
sys.getsizeof(df)
```

```python
# Convert object datatype columns to category
object_columns = ['Attrition','BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
df[object_columns] = df[object_columns].astype('category')

# Convert int64 datatype columns to int32
int64_columns = ['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EmployeeCount', 'EmployeeNumber',
                 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction',
                 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating',
                 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TotalWorkingYears',
                 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
                 'YearsSinceLastPromotion', 'YearsWithCurrManager']
df[int64_columns] = df[int64_columns].astype('int32')
```

```python
sys.getsizeof(df)
```

```python
# deleting EmployeeCount col
df.drop(columns=['EmployeeCount'], inplace=True)
```

```python
# Creating new cols
df.insert(1, 'Attrition1', (df['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)))
```

```python
# rename cols
df.rename(columns={'EmployeeNumber':'EmployeeId'}, inplace=True)
```

```python
# Exporting Data
df.to_csv('GD_attrition.csv', index=False)
```

```python

# Visualizing Data

# Histograms
numeric_columns = df.select_dtypes(include=['int32', 'int64']).columns

for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], bins=20, kde=True, color='skyblue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Box plots
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Attrition', y=column, data=df)
    plt.title(f'Box Plot of {column} by Attrition')
    plt.xlabel('Attrition')
    plt.ylabel(column)
    plt.show()

# Scatter plot (example)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsAtCompany', y='MonthlyIncome', data=df, hue='Attrition')
plt.title('Scatter Plot of Years at Company vs. Monthly Income by Attrition')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.show()
```
